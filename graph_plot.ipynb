```python
import networkx as nx
import matplotlib.pyplot as plt

# 예제 데이터: 논문 간의 인용 관계
citation_data = [
    ("Paper A", "Paper B"),  # A가 B를 인용
    ("Paper A", "Paper C"),  # A가 C를 인용
    ("Paper B", "Paper D"),  # B가 D를 인용
    ("Paper C", "Paper D"),  # C가 D를 인용
    ("Paper C", "Paper E"),  # C가 E를 인용
    ("Paper E", "Paper F"),  # E가 F를 인용
    ("Paper D", "Paper F"),  # D가 F를 인용
    ("Paper F", "Paper G")   # F가 G를 인용
]

# 방향성 그래프 생성
G = nx.DiGraph()
G.add_edges_from(citation_data)

# 그래프 레이아웃 설정
pos = nx.spring_layout(G, seed=42)  # 노드의 위치를 지정

# 시각화
plt.figure(figsize=(10, 7))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=3000,
    node_color="skyblue",
    font_size=10,
    font_weight="bold",
    arrowsize=15,
    edge_color="gray"
)

# 엣지 라벨 추가 (옵션)
edge_labels = {edge: f"cites" for edge in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

plt.title("Research Paper Citation Network", fontsize=14)
plt.show()
```

```python
import plotly.graph_objects as go

# 샘플 데이터
labels = ["Source A", "Source B", "Target X", "Target Y", "Target Z"]
sources = [0, 1, 0, 1, 1]
targets = [2, 2, 3, 3, 4]
values = [8, 4, 2, 4, 6]

# 샌키 다이어그램 생성
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=sources,  # 출발 노드
        target=targets,  # 도착 노드
        value=values     # 연결의 크기
    )
)])

fig.update_layout(title_text="Sankey Diagram Example", font_size=10)
fig.show()
```
